**Answer:**

````python
from pyspark.sql.functions import sqrt

def add_sqrt_column(df, col):
    new_df = df.withColumn("sqrt_" + col, sqrt(col))
    return new_df
````

### Question 3

Write a PySpark code snippet to filter a DataFrame `df` to only include rows where the value in the `column1` column is greater than 10 and the value in the `column2` column contains the substring "foo".

```python
# question code
from pyspark.sql.functions import col

filtered_df = df.filter(...)
```

````python
from pyspark.sql import Row

sample_data = [
    Row(column1=12, column2="foo-bar"),
    Row(column1=5, column2="bar"),
    Row(column1=20, column2="prefixfoo"),
    Row(column1=7, column2="abc")
]

df = spark.createDataFrame(sample_data)
df.show()
````

````python
from pyspark.sql.functions import col

filtered_df = df.filter(
    (col("column1") > 10) &
    col("column2").contains("foo")
)

filtered_df.show()
````
